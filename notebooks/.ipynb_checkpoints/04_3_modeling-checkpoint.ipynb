# Clustering

- `CLASE` (Clase del accidente):
     - 1: Fatal
     - 2: No fatal
     - 3: Sólo daños

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../data/processed/processed_nacional.csv')
```

```python
df.columns
```

```python
df = df.drop(['ANIO', 'CALLE1'], axis=1)
```

```python
scaler = StandardScaler()
```

## Clase

```python
columns = [
    'EDO', 'MES', 'MPIO', 'HORA', 'MINUTOS', 'DIA', 'DIASEMANA',
    'ZONA', 'TIPACCID', 'AUTOMOVIL', 'CAMPASAJ', 'MOTOCICLET', 'BICICLETA',
    'OTROVEHIC', 'CAUSAACCI', 'CAPAROD', 'SEXO', 'EDAD', 'CONDMUERTO',
    'CONDHERIDO', 'PASAMUERTO', 'PASAHERIDO', 'PEATMUERTO', 'PEATHERIDO',
    'CICLMUERTO', 'CICLHERIDO', 'OTROMUERTO', 'OTROHERIDO', 'TOTMUERTOS',
    'TOTHERIDOS', 'LONGITUD', 'LATITUD', 'TRANPUBLICO',
    'VEHICARGA', 'ALIENTOCINT'
]
```

```python
df.loc[:, columns] = scaler.fit_transform(df[columns])
```

```python
X = df.drop('CLASE', axis=1).values
```

```python
df['CLASE'].head()
```

```python
inertia = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(12, 5))

plt.plot(k_values, inertia, marker='o')

plt.show()
```

```python
k_optimo = 4
kmeans = KMeans(n_clusters=k_optimo, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(10, 6))

scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap="viridis", alpha=0.6)
plt.colorbar(scatter, label="Grupos")
plt.show()
```

```python
df['Cluster_Clase'] = clusters
```

```python
df.head()
```

```python
clase_por_grupo = df.groupby('Cluster_Clase')['CLASE'].value_counts(normalize=True) * 100
print("Porcentaje de clase de accidente por grupo:")
print(clase_por_grupo)
```
