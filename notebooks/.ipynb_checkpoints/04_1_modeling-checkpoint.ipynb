# Regression

Para el caso de regresión, a partir de las columnas de `HORA` y `MINUTOS` se crea una variable continua la cual será la variable dependiente a predecir, el tiempo u hora y minutos a la que un accidente ocurrió.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/processed/processed_nacional.csv')
```

```python
df.columns
```

```python
df['MPIO_ID'] = (df['EDO'].astype(str).str.zfill(2) + df['MPIO'].astype(str).str.zfill(3)).astype(int)
```

```python
df[['EDO', 'MPIO', 'MPIO_ID']].head(8)
```

```python
df[['EDO', 'MPIO', 'MPIO_ID']].tail(8)
```

```python
df['CALLE1'].nunique()
```

```python
# Calcular la probabilidad de severidad
df_grouped = df.groupby(['MPIO_ID', 'HORA', 'DIASEMANA']).agg(
    TOTAL_ACCIDENTES=('CLASE', 'count'),
    ACCIDENTES_SEVEROS=('CLASE', lambda x: ((x == 1) | (x == 2)).sum())  # Cuenta fatales + no fatales
).reset_index()

# Calcular la probabilidad de que un accidente sea severo
df_grouped['PROB_SEVERIDAD'] = df_grouped['ACCIDENTES_SEVEROS'] / df_grouped['TOTAL_ACCIDENTES']

# Variables independientes (features)
X = df_grouped[['MPIO_ID', 'HORA', 'DIASEMANA']]
y = df_grouped['PROB_SEVERIDAD']  # Variable dependiente

df_grouped.head(10)
```
